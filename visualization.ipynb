```python
# x = [10]
# y = [20]
x = [2, 4, 6, 8]
y = [7, 14, 21, 28]





plt.grid( color='hotpink', linestyle='dotted', linewidth=0.2,)
plt.subplot(3, 1, 1)
plt.plot(x, y, color='hotpink', marker='X', linestyle='-', linewidth=2, markersize=0.9)
plt.title("my sample plot", )
plt.xlabel("The X AXIS", loc="right")
plt.ylabel("the y axis", loc="top")

x2 = [10, 9, 8, 7, 6]
y2 = [4,8, 16, 20, 25]
plt.subplot(3, 1, 3,)
plt.plot(x2, y2, )
plt.title("my other plot", )

plt.show()
```

```python
x = ["A", "B", "C", "D"]
y = [1,2,3,4]

# plt.barh(x, y)
plt.bar(x, y, width=0.1 , color="green")
plt.title("a graph of letters to number")
plt.show()
```

```python
# x = ["A", "B", "C", "D"]
y = [12, 12 ,40, 40, 40, 43,30,14, 14]

# plt.barh(x, y)
plt.hist( y )
plt.title("a graph of letters to number")
plt.show()
```
